## Mouvement Brownien

### Définition
Soit $(W_t)_{t>=0}$ un processus stochastic. $W_t$ est un mouvement brownien (ou processus de Wiener) si :
+ $W_0 = 0$
+ $\forall$ $0 < t_1 < t_2 <...< t_n $ , les incrments $(W_{t_{i+1}} - W_{t_i})$ sont indépendants et ont une distribution normale, centrée et de variance $(t_{i+1} - t_i)$

```python
import numpy as np

def W_increments(N, h):
    return np.random.normal(0, h, N)

def W_process(N, h):
    incr = W_increments(N-1, h)
    incr = np.insert(incr, 0, 0)
    return incr.cumsum()

```

### Simulations

```python
from matplotlib import pyplot as plt 
N = 1000
h = 0.1
T = [k*h for k in range(N)]

BM0 = W_process(N, h)
BM1 = W_process(N, h)
BM2 = W_process(N, h)
plt.plot([T[0], T[-1]], [0, 0], color = "black") # 0 line 
plt.plot(T, BM0)
plt.plot(T, BM1)
plt.plot(T, BM2)
```
