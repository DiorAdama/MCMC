## Brownian Motion

### Definition
A stochastic process $(W_t)_{t>=0}$ is a brownian motion (Or Wiener process) if it satisfies the following conditions:
+ $ W_0 = 0 $
+ $W_t$ is continuous in t
+ $ \forall $ $0 < t_1 < t_2 <...< t_n $ , the increments $(W_{t_{i+1}} - W_{t_i})$ are independant and normally distributed with mean 0 and variance $(t_{i+1} - t_i)$

```python
import numpy as np

def W_increments(T, N):
    h = T/N
    return np.random.normal(0, np.sqrt(h), N)

def W_process(T, N):
    incr = W_increments(T, N-1)
    incr = np.insert(incr, 0, 0)
    return incr.cumsum()

```

### Simulations

```python
from matplotlib import pyplot as plt 

T = 10
N = 1000
time_steps = np.linspace(0, T, N)

BM0 = W_process(T, N)
BM1 = W_process(T, N)
BM2 = W_process(T, N)
plt.plot([time_steps[0], time_steps[-1]], [0, 0], color = "black") #0 line 
plt.plot(time_steps, BM0)
plt.plot(time_steps, BM1)
plt.plot(time_steps, BM2)
```
